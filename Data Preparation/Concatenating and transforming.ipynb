```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

### Concatenating data

```python
DF_obj = DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj2 = DataFrame(np.arange(15).reshape(5,3))
DF_obj2
```

```python
#join on row
pd.concat([DF_obj, DF_obj2], axis=1)
```

```python
#join on column
pd.concat([DF_obj, DF_obj2], axis=0)
```

### Transforming data
#### Dropping data

```python
#drop rows
DF_obj.drop([0,2])
```

```python
#drop columns
DF_obj.drop([0,2], axis = 1)
```

#### Adding data

```python
series_obj = Series(np.arange(6))
series_obj.name = "ass_variable"
series_obj
```

```python
#join
variable_added = DataFrame.join(DF_obj, series_obj)
variable_added
```

```python
added_datatable = pd.concat([variable_added, variable_added], ignore_index=False)
added_datatable
```

```python
added_datatable = pd.concat([variable_added, variable_added], ignore_index=True)
added_datatable
```

#### Sorting data

```python
DF_sorted = DF_obj.sort_values(by=[5], ascending = False)
DF_sorted
```
