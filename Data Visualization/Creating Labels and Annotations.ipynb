### Labeling plot features
#### The functional method

```python
x = range(1,10)
y= [1,2,3,4,0.5,4,3,2,1]
plt.bar(x,y)
plt.xlabel('your xlabel')
plt.ylabel('your xlabel')
plt
```

```python
z = [1,2,3,4,5]
veh_type = ['bicycle', 'motrobike','car','van','stroller']
plt.pie(z,labels=veh_type)
plt.show()
```

#### The object-oriented method

```python
address = '/content/drive/MyDrive/python-for-data-science-and-machine-learning-essential-training-part-1-3006708-main (1)/data/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = cars.mpg

fig = plt.figure()
ax =fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars')
ax.set_xlabel('cars names')
ax.set_ylabel('miles/gal')
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show
```

#### The object-oriented method

```python
fig = plt.figure()
ax =fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars')
ax.set_xlabel('cars names')
ax.set_ylabel('miles/gal')
ax.legend(loc = 'best')
```

### Annotating your plot

```python
fig = plt.figure()
ax =fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars')
ax.set_xlabel('cars names')
ax.set_ylabel('miles/gal')

ax.set_ylim([0,45])
ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35), arrowprops=dict(facecolor='black', shrink=.05))
ax.legend(loc = 'best')
```
