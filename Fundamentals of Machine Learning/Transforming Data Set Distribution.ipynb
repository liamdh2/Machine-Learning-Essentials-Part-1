## Transforming Dataset Distributions

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, scale
```

```python
# prompt: mount drive

from google.colab import drive
drive.mount('/content/drive')
address = '/content/drive/MyDrive/python-for-data-science-and-machine-learning-essential-training-part-1-3006708-main (1)/data/mtcars.csv'
```

```python
dataset = pd.read_csv(address)
dataset.head()
```

```python
plt.plot(dataset[['mpg']])
```

### Normalization

```python
MinMaxScaler=MinMaxScaler()
MinMaxScaler.fit(dataset[['mpg']])
```

```python
scaled = MinMaxScaler.transform(dataset[['mpg']])
plt.plot(scaled)
```

### Standardization

```python
st_scalar = scale(dataset[['mpg']])
plt.plot(st_scalar)
```
